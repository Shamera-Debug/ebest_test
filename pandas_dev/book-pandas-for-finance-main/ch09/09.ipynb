```python
import numpy as np
import talib
import pandas as pd
```

```python
stock_data = pd.read_excel('../ch08/data/samsung.xlsx')
stock_data.head()
```

```python
stock_data.set_index('일자',inplace=True)
stock_data.sort_index(inplace=True)
stock_data.head()
```

```python
sma5 = talib.SMA(stock_data['종가'], timeperiod=5)
sma20 = talib.SMA(stock_data['종가'], timeperiod=20)
sma60 = talib.SMA(stock_data['종가'], timeperiod=60)
```

```python
sma5
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.plot(stock_data['종가'])
plt.plot(sma5, label='5-day SMA')
plt.plot(sma20, label='20-day SMA')
plt.plot(sma60, label='60-day SMA')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
# 스토케스틱 - 오실레이터 : 
so = talib.STOCH(
      stock_data['고가']
    , stock_data['저가']
    , stock_data['종가']
    , fastk_period=14
    , slowk_period=3
    , slowk_matype=0
    , slowd_period=3
    , slowd_matype=0
)
```

```python
so
```

```python
# RSI : 상태강도지수 : 매수매도 타이밍
rsi = talib.RSI(stock_data['종가'], timeperiod=14)
rsi
```
