## 7.3 칼럼 시프트

```python
import pandas as pd
df = pd.read_excel("data/data_5403_20231205.xlsx" , index_col=0)
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df
```

```python

df["거래량"].shift(1)
```

```python
df["전일거래량"] = df["거래량"].shift(1)
df[ ['거래량', '전일거래량'] ]
```

```python
# 전일보다 거래량이 많은 일자 추출
df["전일거래량"] = df["거래량"].shift(1)
cond = df["거래량"] > df["전일거래량"] # boolean
df[cond]
```

```python
print("거래량 상승일:", len(df[cond]))
print("전체 영업일:", len(df))
```

```python
df['거래량'].diff( )
```

```python
cond = df['거래량'].diff() > 0
print('거래량 상승일 ',len(df[cond]))
```

```python
yeild = df['종가'] / df['종가'].shift(6) # 6일전 종가
cond = yeild >= 1.03 
df[cond]
```

```python
cond.shift(1)
```

```python
cond.shift(1)
```

```python
cond_modified = cond.shift(1).fillna(False)
s = df.loc[cond_modified, '종가'] / df.loc[cond_modified, '시가']
print(s.cumprod().iloc[-1])
```

## 7.6 수익률 계산하기

```python
import pandas as pd

data = {'삼성전자': [52200, 52300, 52900, 52000, 51700], 
           'LG전자': [68200, 67800, 68800, 67500, 66300]}
df = pd.DataFrame(data=data)
df.pct_change()
```

```python
df / df.shift(2) - 1
```

```python
df.pct_change(periods=2)
```

```python
yeild = df.pct_change(periods=2) + 1
yeild.cumprod()
```

```python
df = pd.read_excel("data/ss_ex_1.xlsx", index_col=0, usecols=[0, 1, 4])
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df.head()
```

```python
# 각 분기의 시가의 처음값 = > 데이프레임으로 변환
df_quarter = df['시가'].resample('q').first().to_frame()
df_quarter
```

```python
df['시가'].groupby(pd.Grouper(freq='q')).first().to_frame()
```

```python
df_quarter.index.quarter
```

```python
df_quarter['quarter'] = df_quarter.index.quarter
df['quarter'] = df.index.quarter
```

```python
df_quarter
```

```python
df.head()
```

```python
df_daily = df[['종가', 'quarter']].reset_index()
df_daily
```

```python

r = pd.merge(left=df_daily, right=df_quarter, on='quarter')
r
```

```python
r['수익률'] = r['종가'] / r['시가']
r = r.set_index(['quarter', '일자'])
r
# 각 분기의 시가 대비 일일 상승률(종가)
```
