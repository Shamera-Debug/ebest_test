```python
def cate(df):
    if df['Open'] < df['Close']:
        return '상승'
    elif df['Open'] == df['Close']:
        return '보합'
    else:
        return '하락'
```

```python
kospi['cate'] = kospi.map(cate, )
kospi.head()
```

```python
upward     = kospi['Open']  < kospi['Close'] # 종가가 시작가 큰날, 비교연산자 => boolean
stationary = kospi['Open'] == kospi['Close'] # 같은날
downward   = kospi['Open']  > kospi['Close'] # 작은날           

kospi.loc[upward,     'category'] = "상승"
kospi.loc[stationary, 'category'] = "보합"
kospi.loc[downward,   'category'] = "하락"

how = {
    "category": len
}
data = kospi.groupby('category').agg(how)
data
```
